```python
import torch
import pandas as pd
import numpy as np
```

```python
unique_dataid = torch.load("../data/Pecan Street Smart Meter Data (large) (tensor)/data_id.pt")
unique_dataid = np.array(unique_dataid)
tensor_grid = torch.load("../data/Pecan Street Smart Meter Data (large) (tensor)/loads_raw.pt")
```

```python
profile = pd.read_csv('../data/Pecan Street Smart Meter Data (large) (raw)/metadata.csv')
profile = profile.drop(profile.index[0])

profile['dataid'] = profile['dataid'].astype(int)
profile_filter = profile[profile['dataid'].isin(unique_dataid)].copy()


profile_filter['dataid'] = pd.Categorical(
    profile_filter['dataid'], categories=unique_dataid, ordered=True)


# Sort by the 'dataid' column to get the rows in the same order as unique_dataid
profile_filter = profile_filter.sort_values('dataid')

# Reset the index of the resulting DataFrame
profile_filter = profile_filter.reset_index(drop=True)
profile_filter
```

```python
pv_profile = profile_filter[["total_amount_of_pv", "amount_of_south_facing_pv",
                             "amount_of_west_facing_pv", "amount_of_east_facing_pv"]]
pv_profile = pv_profile.fillna(0)
pv_profile
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import torch  # Make sure to import torch

# Creating a DataFrame with dates for a non-leap year
start_date = '2018-01-01'
end_date = '2018-12-31'
dates = pd.date_range(start=start_date, end=end_date, freq='D')
te = pd.DataFrame({'Date': dates})

# Extracting month, day of the month, and day of the week
te['Month'] = te['Date'].dt.month
te['Day'] = te['Date'].dt.day
te['Weekday'] = te['Date'].dt.weekday + 1  # Making Monday = 1, Sunday = 7

# One-hot encoding for Month, Day, and Weekday
encoder = OneHotEncoder(sparse_output=False)  # Use sparse_output for recent versions
encoded_features = encoder.fit_transform(te[['Month', 'Weekday', 'Day']])

# Converting the encoded features back to a DataFrame for better visualization
encoded_te = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(['Month', 'Weekday', 'Day']))

# Converting to PyTorch tensor_full
encoded_te_tensor = torch.tensor(encoded_te.values, dtype=torch.float32)  # Use .values for NumPy array
encoded_te_tensor = encoded_te_tensor.unsqueeze(0).repeat(len(unique_dataid), 1, 1)  # Adjusting shape as needed

# Output the shape of the tensor_full
print(encoded_te_tensor.shape)
```

```python
# profile = pd.concat([pv_profile, use1, use2], axis=1)
profile_tensor = torch.tensor(np.array(pv_profile, dtype=np.float32))
profile_tensor = torch.concat([profile_tensor, torch.eye(len(profile_tensor))], dim=1)
profile_tensor = profile_tensor.unsqueeze(1).repeat(1, encoded_te_tensor.shape[1], 1)
```
